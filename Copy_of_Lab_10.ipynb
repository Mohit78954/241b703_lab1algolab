# Warshall algorithm to find reachability between all pairs of vertices in a graph.
Problem Description:
Understand transitive closure of a graph. Use Python to compute reachability using matrix method.

#Theory
Warshall algorithm finds whether a path exists between every pair of vertices. It updates the adjacency matrix step by step. If there is a path from i to j through k, then the entry becomes 1.

## Algorithm
1. Represent graph using adjacency matrix.
2. For each intermediate vertex k
3. Check all pairs i and j
4. Update value if path exists through k
4. Repeat for all vertices


```python
def warshall(graph):
    n = len(graph)


    reach = [row[:] for row in graph]

    for k in range(n):
        for i in range(n):
            for j in range(n):
                reach[i][j] = reach[i][j] or (reach[i][k] and reach[k][j])

    return reach


def floyd(graph):
    n = len(graph)
    INF = 999999

    dist = []

    for i in range(n):
        row = []
        for j in range(n):
            if i == j:
                row.append(0)
            elif graph[i][j] != 0:
                row.append(graph[i][j])
            else:
                row.append(INF)
        dist.append(row)

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]

    return dist
```

```python
n = 5
graph = []

print("Enter the adjacency matrix for a graph with 5 vertices:")
print("Enter 5 rows, each containing 5 values (0 or 1):")

for i in range(n):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    while len(row) != n:
        print("Please enter exactly 5 values.")
        row = list(map(int, input(f"Row {i+1}: ").split()))
    graph.append(row)


warshall_result = warshall(graph)

print("\nTransitive Closure Matrix using Warshall Algorithm:")
for row in warshall_result:
    print(row)


floyd_result = floyd(graph)

print("\nShortest Path Matrix using Floyd Algorithm:")
for i in range(n):
    for j in range(n):
        if floyd_result[i][j] == 999999:
            print("INF", end="\t")
        else:
            print(floyd_result[i][j], end="\t")
    print()


print("\nPath Existence between vertices:")
for i in range(n):
    for j in range(n):
        if warshall_result[i][j]:
            print(f"Path exists from Vertex {i} to Vertex {j}")
        else:
            print(f"No path exists from Vertex {i} to Vertex {j}")

# Comparison
print("\nComparison between Warshall and Floyd Algorithm:")
print("1. Warshall Algorithm:")
print("   - Used to find whether a path exists between two vertices.")
print("   - Output is Transitive Closure Matrix (True/False or 1/0).")

print("2. Floyd Algorithm:")
print("   - Used to find the shortest path between all pairs of vertices.")
print("   - Output is Shortest Distance Matrix.")

print("\nIn short:")
print("Warshall -> checks path existence")
print("Floyd    -> finds shortest path")
```

## Find shortest paths between all pairs of vertices using Floyd Warshall algorithm.

#Understand all pairs shortest path problem. Implement dynamic programming approach in Python.

* Given a weighted graph with positive or negative edge weights, find
shortest distances between every pair of vertices.

# Theory
Floyd Warshall algorithm computes shortest paths between all pairs of nodes. It updates the distance matrix by checking if a shorter path exists through an intermediate vertex. Works with negative weights but fails with negative cycles.

Algorithm
1. Represent graph using distance matrix.
2. Initialize matrix with given weights.
3. Set distance of node to itself as 0.
4. For each intermediate vertex k
5. Update distance[i][j] = min(distance[i][j], distance[i][k] + distance[k][j])
6. Repeat for all vertices

```python
def floyd_warshall(graph):
    n = len(graph)
    dist = []
    for i in range(n):
        dist.append(graph[i][:])


    nxt = [[-1] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            if graph[i][j] != INF and i != j:
                nxt[i][j] = j


    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] != INF and dist[k][j] != INF:
                    if dist[i][k] + dist[k][j] < dist[i][j]:
                        dist[i][j] = dist[i][k] + dist[k][j]
                        nxt[i][j] = nxt[i][k]

    return dist, nxt


def detect_negative_cycle(dist):
    n = len(dist)
    for i in range(n):
        if dist[i][i] < 0:
            return True
    return False


def get_path(nxt, u, v):
    if nxt[u][v] == -1:
        return []

    path = [u]
    while u != v:
        u = nxt[u][v]
        path.append(u)
    return path


def dijkstra(graph, src):
    n = len(graph)
    visited = [False] * n
    dist = [INF] * n
    dist[src] = 0

    for _ in range(n):
        u = -1
        min_dist = INF

        for i in range(n):
            if not visited[i] and dist[i] < min_dist:
                min_dist = dist[i]
                u = i

        if u == -1:
            break

        visited[u] = True

        for v in range(n):
            if graph[u][v] != INF and not visited[v]:
                if dist[u] + graph[u][v] < dist[v]:
                    dist[v] = dist[u] + graph[u][v]

    return dist


INF = float('inf')
```

```python
n = int(input("Enter number of vertices: "))

print("Enter the adjacency matrix:")
print("Use -1 for no direct edge.")

graph = []
for i in range(n):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    while len(row) != n:
        print(f"Please enter exactly {n} values.")
        row = list(map(int, input(f"Row {i+1}: ").split()))

    new_row = []
    for j in range(n):
        if row[j] == -1:
            new_row.append(INF)
        else:
            new_row.append(row[j])
    graph.append(new_row)


dist, nxt = floyd_warshall(graph)

print("\nShortest distance matrix:")
for i in range(n):
    for j in range(n):
        if dist[i][j] == INF:
            print("INF", end="\t")
        else:
            print(dist[i][j], end="\t")
    print()


if detect_negative_cycle(dist):
    print("\nGraph contains a negative cycle.")
else:
    print("\nGraph does not contain any negative cycle.")


u = int(input("\nEnter source vertex: "))
v = int(input("Enter destination vertex: "))

path = get_path(nxt, u, v)

if len(path) == 0:
    print(f"No path exists from {u} to {v}.")
else:
    print(f"Shortest distance from {u} to {v} is: {dist[u][v]}")
    print("Shortest path is:", " -> ".join(map(str, path)))


print("\nComparison with Dijkstra Algorithm:")
print("1. Floyd-Warshall Algorithm:")
print("   - Finds shortest paths between all pairs of vertices.")
print("   - Works for positive and negative edge weights.")
print("   - Can detect negative cycles.")
print("   - Time Complexity: O(n^3)")

print("\n2. Dijkstra Algorithm:")
print("   - Finds shortest path from one source to all other vertices.")
print("   - Works only for non-negative weights.")
print("   - Cannot detect negative cycles.")
print("   - Time Complexity: O(n^2) for simple version")

choice = input("\nDo you want to run Dijkstra from a source vertex? (yes/no): ")

if choice.lower() == "yes":
    src = int(input("Enter source vertex for Dijkstra: "))
    dijkstra_dist = dijkstra(graph, src)

    print(f"\nShortest distances from source {src} using Dijkstra:")
    for i in range(n):
        if dijkstra_dist[i] == INF:
            print(f"{src} -> {i} = INF")
        else:
            print(f"{src} -> {i} = {dijkstra_dist[i]}")
```
