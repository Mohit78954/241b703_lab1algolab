```python
# Exercise 1
import timeit
def func1(a, b):
  result = 0
  for i in range(0, b):
    result += a
  return result
```

```python
# Exercise 2
import timeit
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
```

```python
# Exercise 4
import timeit
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
start_time=timeit.default_timer()
x=print(func4([1,2,3,4],[2,4,6,8]))
end_time=timeit.default_timer()
print('>> elapsed time',(end_time-start_time)*1000)
#Here in this we perform binary search in func4 which divide the array in two parts so that it will take less time to find the target value in a optimize way which will make our algorithm more efficient,
```

```python
# Exercise 5
import timeit
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
start_time=timeit.default_timer()
x=print(func5(16))
end_time=timeit.default_timer()
print('>> elapsed time',(end_time-start_time)*1000)
#this fuction performs the repeation of comparing and addition until the multiplication of the guess value become equal to the input value that we wil be giving.
```

```python
# Exercise 6
import timeit
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
start_time=timeit.default_timer()
x=print(func8([3,1,2,7,8,6]))
end_time=timeit.default_timer()
print('>> elapsed time',(end_time-start_time)*1000)
```

```python
# Exercise 7
import timeit
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
# write a function named as sum_arr to find sum of elements in given array
import timeit
def sum_arr(arr):
  sum=0
  for i in range(len(arr)):
    sum+=arr[i]
  return sum
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
import timeit
def Counting_EvenOdd(arr):
  count=0
  for i in range(len(arr)):
    if(arr[i]%2==0):
      count+=1
  return count

```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr,l,h,target):
  mid=(l+h)//2
  if h<l:
    return -1
  elif target== arr[mid]:
    return mid
  elif target<arr[mid]:
    return Binary_Search(arr,l,mid-1,target)
  else:
    return Binary_Search(arr,mid+1,h,target)

```

```python
# call the function Binary_Search for array
arr = [1, 3, 5, 7, 9,15,16,14,45]
start_time=timeit.default_timer()
print('this target value found at index ',Binary_Search(arr,0,len(arr),7))
end_time=timeit.default_timer()
print('>> elapsed time',(end_time-start_time)*1000)
```
